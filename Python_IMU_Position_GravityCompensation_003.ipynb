### eigentliche Zeit-Differenz-Berechnung

```python
import numpy as np
from datetime import datetime

time = df['Time'].values  # Zeitspalte falls vorhanden

# Dein NumPy-Array mit Zeitstempeln als Strings
# time = np.array([' 18:06:05.104', ' 18:06:05.247', ' 18:06:05.304', ' 18:06:05.403'])

# Entferne führende und nachfolgende Leerzeichen aus der 'Time'-Spalte
df['Time'] = df['Time'].str.strip()
 
# Konvertiere die 'Time'-Spalte in das Datetime-Format
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S.%f')

# Zeitdifferenzen berechnen und in einer neuen Spalte speichern
df['Delta_t'] = df['Time'].diff().dt.total_seconds()
 
# print(df['Delta_t'])
```

```python
import numpy as np
import pandas as pd

# Erdbeschleunigung
g = 9.81

# Funktion zur Berechnung der Beschleunigungskomponenten
def calculate_acceleration(row):
    roll_rad = np.radians(row['Angle X(°)'])
    pitch_rad = np.radians(row['Angle Y(°)'])
    
    a_x = g * np.sin(pitch_rad)
    a_y = g * np.sin(roll_rad)
    a_z = g * np.cos(pitch_rad) * np.cos(roll_rad)
    
    return pd.Series([a_x, a_y, a_z])

# Anwenden der Funktion auf den DataFrame und Hinzufügen der Spalten
df[['Acceleration X (m/s²)', 'Acceleration Y (m/s²)', 'Acceleration Z (m/s²)']] = df.apply(calculate_acceleration, axis=1)

# Berechnung der Summe der Beschleunigungskomponenten pro Zeile
df['Acceleration Sum (m/s²)'] = df[['Acceleration X (m/s²)', 'Acceleration Y (m/s²)', 'Acceleration Z (m/s²)']].sum(axis=1)

# Ausgabe des erweiterten DataFrames
df.head(5)
```
